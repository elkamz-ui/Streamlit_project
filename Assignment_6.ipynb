```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Pandas Project.csv")
type(df)
```

```python
top_7_sales_person = df["Sales Person"].value_counts().head(10).index.to_list()
top_7_sales_person
# top_7_products = df["Product"].value_counts().head(7).index.to_list()
# top_7_products
```

```python
df["Amount"] = pd.to_numeric(df["Amount"].replace({r'[\$,]': ''}, regex=True))
df.info()
```

```python
revenue_per_salesperson = df.groupby("Sales Person")["Amount"].sum()
revenue_per_salesperson.sort_values(ascending=False)
```

```python
boxes_shipped_per_product =df.groupby("Product")["Boxes Shipped"].sum()
boxes_shipped_per_product.sort_values(ascending=False)
```

```python
df["Date"] = pd.to_datetime(df["Date"], format="%d/%m/%Y")
df["Month_name"] = df["Date"].dt.month_name()
df.info()
```

```python
total_sales_per_month = df.groupby("Month_name")["Month_name"].count()
total_sales_per_month
```

```python
df['Revenue_per_box'] = df['Amount'] / df['Boxes Shipped']
avg_revenue_per_box = df['Revenue_per_box'].mean()
# df['Revenue_per_box']
df[['Amount', 'Boxes Shipped', 'Revenue_per_box']].head(20)
# avg_revenue_per_box
```

```python
total_revenue_per_revenue = df.groupby("Country")["Amount"].sum()
total_revenue_per_revenue.sort_values(ascending=False)
```

```python
revenue_per_product = df.groupby("Product")["Amount"].sum()
revenue_per_product.sort_values(ascending=False)
```
